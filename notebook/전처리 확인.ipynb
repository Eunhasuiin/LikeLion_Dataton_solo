```python
def extract(temp, match_quantity, select_list):
    match_df = pd.DataFrame()
    #game_basic_data
    match_df['matchId'] = [temp[x]["info"]['gameId'] for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']
    match_df['blue/red'] = [True if y < 5 else False for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']
    match_df['gameDuration'] = [temp[x]["info"]['gameDuration'] for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']
    match_df['puuid'] = [temp[x]["info"]['participants'][y]['puuid'] for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']
    match_df['win'] = [temp[x]["info"]['participants'][y]['win'] for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']
    match_df['is_Surrender_aft_25m'] = [temp[x]["info"]['participants'][y]['gameEndedInSurrender'] for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']
    match_df['is_Early_Surrender'] = [temp[x]["info"]['participants'][y]['gameEndedInEarlySurrender'] for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']
    match_df['is_replay'] = [True if temp[x]['info']['gameDuration'] < 270 else False for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']
        
    #personal_basic_data
    match_df['teamPosition'] = [temp[x]["info"]['participants'][y]['teamPosition'] for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']
    match_df['championName'] = [temp[x]["info"]['participants'][y]['championName'] for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']
        
    for select_ in select_list:
        match_df[select_] = [temp[x]["info"]['participants'][y][select_] for x in range(match_quantity) for y in range(10) if temp[x]['info']['endOfGameResult'] == 'GameComplete']

    print(f"{len(match_df)}개 row")
    return match_df

#----------------------------------------Main_Stream----------------------------------------------------------
tier_list=['diamond','emerald']
rank_list = ['i','ii','iii','iv']
date_list=['2025-07-14_to_15','2025-07-20_to_23']
essential_list=['matchId','blue/red','gameDuration','puuid','win','is_Surrender_aft_25m','is_Early_Surrender','is_replay','teamPosition','championName']

#수정은 이 부분만 해주시면 됩니다.
ping_list = ['commandPings', 'getBackPings' , 'holdPings', 'onMyWayPings', 'allInPings', 'assistMePings' ,'needVisionPings', 'enemyMissingPings','enemyVisionPings']

for tier in tier_list:
    output_df = pd.DataFrame(columns = essential_list + ping_list)
    for rank in rank_list:
        for date_ in date_list: 
            with open(f"{input_dir}{tier}_{rank}_matches_{date_}.json",'rb') as f: temp=json.load(f)    
            
            print(f"{tier}_{rank}_{date_} 파일 처리 시작")
            match_quantity = len(temp)
            match_df = extract(temp, match_quantity, ping_list)
            output_df = pd.concat([output_df, match_df],axis=0)
            print(f"{tier}_{rank}_{date_} 파일 처리 끝",end='\n\n')
        
    print(f'{tier} 완료')
    output_df.to_csv(f'{tier}_ping.csv',index_label=False)
    print(f'{tier}_ping.csv 제출 완료',end = '\n\n\n')
```
